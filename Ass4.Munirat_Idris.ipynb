```python
### Question 2 part 2
##Code for Standard Matrix Multiplication using random variables:
import time
import random

def matrix_multiply(matrix1, matrix2):
    """
    Perform matrix multiplication using nested loops.
    """
    m = len(matrix1)
    n = len(matrix1[0])
    p = len(matrix2[0])
    
    result = [[0 for _ in range(p)] for _ in range(m)]
    
    for i in range(m):
        for j in range(p):
            for k in range(n):
                result[i][j] += matrix1[i][k] * matrix2[k][j]
    
    return result

# Function to generate a random n x n matrix
def generate_random_matrix(n):
    return [[random.randint(1, 100) for _ in range(n)] for _ in range(n)]

# Vary n and measure execution time
n_values = [10, 50, 100, 200, 300]
execution_times = []

for n in n_values:
    matrix1 = generate_random_matrix(n)
    matrix2 = generate_random_matrix(n)
    
    start_time = time.time()
    result = matrix_multiply(matrix1, matrix2)
    end_time = time.time()
    
    execution_time = end_time - start_time
    execution_times.append(execution_time)

# Plot execution time vs n
import matplotlib.pyplot as plt

plt.plot(n_values, execution_times, marker='o')
plt.title('Execution Time vs Matrix Size (Standard Matrix Multiplication)')
plt.xlabel('Matrix Size (n x n)')
plt.ylabel('Execution Time (seconds)')
plt.grid(True)
plt.show()

### Code for NumPy Matrix Multiplication:

import time
import numpy as np

# Function to generate a random n x n matrix using NumPy
def generate_random_matrix_numpy(n):
    return np.random.randint(1, 100, size=(n, n))

# Vary n and measure execution time
n_values = [10, 50, 100, 200, 300]
execution_times_numpy = []

for n in n_values:
    matrix1 = generate_random_matrix_numpy(n)
    matrix2 = generate_random_matrix_numpy(n)
    
    start_time = time.time()
    result = np.dot(matrix1, matrix2)
    end_time = time.time()
    
    execution_time = end_time - start_time
    execution_times_numpy.append(execution_time)

# Plot execution time vs n
import matplotlib.pyplot as plt

plt.plot(n_values, execution_times_numpy, marker='o')
plt.title('Execution Time vs Matrix Size (NumPy Matrix Multiplication)')
plt.xlabel('Matrix Size (n x n)')
plt.ylabel('Execution Time (seconds)')
plt.grid(True)
plt.show()
```

Interpretation of Results:
The execution time increases as the size of the matrix (n) increases for both the standard matrix multiplication approach and NumPy matrix multiplication.
NumPy matrix multiplication is significantly faster compared to the standard nested loop approach, especially for larger matrix sizes. This is because NumPy utilizes highly optimized C and Fortran libraries under the hood, resulting in efficient matrix operations.
As expected, the execution time increases non-linearly with the matrix size (n) for both methods. However, the execution time is much faster for NumPy compared to the standard approach, demonstrating the advantage of using optimized libraries for matrix operations.


```python
#Question 3
import numpy as np 
from scipy.linalg import svd 
# Define matrix A
A = np.array([
[41, 92, 53],
[84, 75, 36],
[87, 38, 29]    
])

# Perform SVD decomposition
U, s, V = svd(A) 

# Print the results 
print("U:") 
print(U) 
print("s:") 
print(s) 
print("V:") 
print(V) 


# Reconstruct A using the SVD components
reconstructed_A = np.dot(U, np.dot(np.diag(s), V))

# Print original matrix A and reconstructed A
print("Original matrix A:\n", A)
print("\nReconstructed matrix A:\n", reconstructed_A)


```

In this code, I used the `numpy` library but the `svd` function is from the `scipy.linalg` module. The input matrix `A` is a 3x3 matrix. Then, I used the `svd` function to perform the SVD on matrix `A`, which returns three matrices: `U` (containing the left singular vectors), `s` (containing the singular values as a 1-D array), and `V` (containing the right singular vectors). Note that this code uses the `scipy.linalg.svd` function instead of `numpy.linalg.svd`. The `scipy.linalg.svd` function provides a more robust implementation of SVD, but the usage is similar to `numpy.linalg.svd`.
